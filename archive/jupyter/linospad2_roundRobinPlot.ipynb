```python
## imports
import numpy as np
from matplotlib import pyplot as plt
from struct import unpack
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
from scipy.stats import norm
from scipy.optimize import curve_fit
from astropy import modeling
%matplotlib widget
```

```python
## Compute histogram of valid time stamps per pixel
# Masking out pixels 15 and 93

beam1 = np.arange(0,256,1)
beam2 = -100*np.ones(256)

beam1_spots=[87, 135, 135,135, 36, 104, 80, 80, 80, 36, 36 ,164, 133, 164, 164, 199, 33, 29,38,36,41,21,104,72]
beam2_spots=[219, 171, 190, 219, 219, 219, 246, 201, 176, 246, 224, 176, 176, 191,192, 206, 78,181,82,170,58,43,116,85] 

for i in range(len(beam1_spots)):
    beam1[i] = beam1_spots[i]
    beam2[i] = beam2_spots[i]
    beam1[i+len(beam1_spots)] = beam2_spots[i]
    beam2[i+len(beam1_spots)] = beam1_spots[i]

fig, ax = plt.subplots(1,1)

plt.rcParams.update({"font.size": 18})



plt.box(bool(1))
plt.grid(color='black', linestyle='-', linewidth=1)
plt.grid(True)
plt.subplots_adjust(left=0.15)
fig.set_size_inches(8, 8)

ax.tick_params(which="both", width=2, direction="in")
ax.tick_params(which="major", length=7, direction="in")
ax.tick_params(which="minor", length=4, direction="in")
ax.yaxis.set_ticks_position("both")
ax.xaxis.set_ticks_position("both")
plt.xlabel("#1 beam position [pixel]")
plt.ylabel("#2 beam position [pixel]")

for axis in ["top", "bottom", "left", "right"]:
    ax.spines[axis].set_linewidth(2)

ax.set_title("Coincidence ");    
ax.scatter(beam1, beam2,50);
ax.xaxis.set_major_locator(MultipleLocator(64))
ax.yaxis.set_major_locator(MultipleLocator(64))
ax.set_xlim(0, 256)
ax.set_ylim(0, 256)
```
